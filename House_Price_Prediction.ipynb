**Import the Libraries**

```
 import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

 **Import the Dataset**

```
 from sklearn.datasets import load_boston
boston = load_boston()
```

```
 boston.keys()
```

```
 print (boston.DESCR)
```

```
 df = pd.DataFrame(boston.data)
```

```
 df.columns = boston.feature_names
```

 **Exploratory Data Analysis**

Di sini, kita akan melihat beberapa hubungan antara variabel Independen dan Variabel Dependen.

**1. Crime Rate & PRICE**

```
 plt.figure(figsize=(10,5))
sns.regplot(x=df.CRIM, y=df.PRICE)
plt.title('Relationship between CRIME Rate and Price of the Houses')
plt.xlabel('Per capita crime rate by town (CRIM)')
plt.ylabel('Housing Prices')
```

 **Negative Linear Relation:** Di sini, kita dapat melihat bahwa, ketika Tingkat Kejahatan meningkat, Harga Rumah menurun.

**2. Average Number of Rooms & PRICE**

```
 plt.figure(figsize=(10,5))
sns.regplot(x=df.RM, y=df.PRICE)
plt.title('Relationship between Avg. No. of Rooms and Price of the Houses')
plt.xlabel('Average number of rooms per dwelling(RM)')
plt.ylabel('Housing Prices')
```

**Positive Linear Relation:** Di sini, kita dapat melihat bahwa, dengan meningkatnya Jumlah Kamar, Harga Rumah meningkat.

**3. Pupil-Teacher Ratio in Locality & PRICE**

```
 plt.figure(figsize=(10,5))
sns.regplot(x=df.PTRATIO, y=df.PRICE)
plt.title('Relationship between Pupil-Teacher Ratio and Price of the Houses')
plt.xlabel('Pupil-Teacher ratio by town(PTRATIO)')
plt.ylabel('Housing Prices')
```

**Negative Linear Relation:** Di sini, kita dapat melihat bahwa, ketika rasio Jumlah Murid-Guru meningkat, Harga Rumah turun.

 **4. Land Zone Proportion & PRICE**

```
plt.figure(figsize=(10,5))
sns.regplot(x=df.ZN, y=df.PRICE)
plt.title('Relationship between Land Zone Proportion and Price of the Houses')
plt.xlabel('proportion of residential land zoned for lots over 25,000 sq.ft.(ZN)')
plt.ylabel('Housing Prices')
```

**Positive Lineaar Relation:** Di sini, kita dapat melihat bahwa, seiring dengan meningkatnya proporsi tanah perumahan yang dikategorikan untuk kavling di atas 25.000 sq.ft. (ZN), Harga Rumah meningkat.

 **5. Non-retail Business acres per town Proportion & PRICE**

```
plt.figure(figsize=(10,5))
sns.regplot(x=df.INDUS, y=df.PRICE)
plt.title('Relationship between Non-retail Business Proportion and Price of the Houses')
plt.xlabel('proportion of non-retail business acres per town(INDUS)')
plt.ylabel('Housing Prices')
```

**Negative Linear Relation:** Di sini, kita dapat melihat bahwa, ketika proporsi hektar bisnis non-ritel per kota meningkat, Harga Rumah menurun.

**6. Nitric Oxides Concentration & PRICE**

```
plt.figure(figsize=(10,5))
sns.regplot(x=df.NOX, y=df.PRICE)
plt.title('Relationship between Nitric Oxides Concentration and Price of the Houses')
plt.xlabel('Nitric Oxides Concentration (parts per 10 million(NOX)')
plt.ylabel('Housing Prices')
```

**Negative Linear Relation:** Di sini, kita dapat melihat bahwa dengan meningkatnya Konsentrasi Oksida Nitrat, Harga Rumah menurun.
